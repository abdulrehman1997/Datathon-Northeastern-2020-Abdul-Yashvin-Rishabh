```python
import numpy as np
import pandas as pd
```

```python
covid_data = pd.read_csv('../../covid-19-county-level-data.csv')
mobility = pd.read_csv('../../us-mobility.csv')
```

```python
covid_data = covid_data[covid_data.state == 'Massachusetts']
mobility = mobility[mobility.state == 'Massachusetts']
```

```python
mobility['county'] = mobility.county.apply(lambda x: str(x).split()[0])
mobility = mobility.drop(['Unnamed: 0','state'], axis=1)
covid_data = covid_data.drop(['Unnamed: 0','fips','state'],axis=1)
```

```python
df = pd.merge(covid_data, mobility,  how='left', left_on=['date','county'], right_on = ['date','county',])
df = df[df['county']!='Total']
df = df[df['county']!='Unknown']
```

```python
df_no_na = df.dropna()
```

```python
df["retail_and_recreation"] = df.groupby("county")["retail_and_recreation"].transform(lambda x: x.fillna(x.median()))
df["grocery_and_pharmacy"] = df.groupby("county")["grocery_and_pharmacy"].transform(lambda x: x.fillna(x.median()))
df["parks"] = df.groupby("county")["parks"].transform(lambda x: x.fillna(x.median()))
df["transit_stations"] = df.groupby("county")["transit_stations"].transform(lambda x: x.fillna(x.median()))
df["workplaces"] = df.groupby("county")["workplaces"].transform(lambda x: x.fillna(x.median()))
df["residential"] = df.groupby("county")["residential"].transform(lambda x: x.fillna(x.median()))
df = df.fillna(0)
```
